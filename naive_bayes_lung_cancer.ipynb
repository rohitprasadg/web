```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("/content/survey lung cancer.csv")
```

```python
df.isnull().sum()
```

```python
X = df.drop('LUNG_CANCER', axis = 1)
y = df['LUNG_CANCER']
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
df['GENDER']=le.fit_transform(df['GENDER'])
df['LUNG_CANCER']=le.fit_transform(df['LUNG_CANCER'])
```

```python
df_new=df.drop(columns=['GENDER','AGE', 'SMOKING', 'SHORTNESS OF BREATH'])
df_new
```

```python
X=df_new.drop("LUNG_CANCER", axis = 1)
y=df_new['LUNG_CANCER']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb_model = GaussianNB()
gnb_model.fit(X_train, y_train)
```

```python
y_gnb_pred= gnb_model.predict(X_test)
y_gnb_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix,  classification_report
gnb_cr=classification_report(y_test, y_gnb_pred)
gnb_cm=confusion_matrix(y_test, y_gnb_pred)
gnb_ac=accuracy_score(y_test, y_gnb_pred)
print("Classification report:",gnb_cr)
print("Confusion matrix:\n",gnb_cm)
print("\nAccuracy:",gnb_ac)
```
